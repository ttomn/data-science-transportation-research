```python
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport
import seaborn as sns
import folium
```

```python
df_summons = pd.read_csv("datasets/NYPD_B_Summons__Year_to_Date_.csv")
df_summons
```

```python
df_collisions = pd.read_csv("datasets/Motor_Vehicle_Collisions_-_Crashes.csv")
df_collisions
```

```python
america_centre_coords = [40.765792264679234, -73.97331919385795]
m_1 = folium.Map(location=america_centre_coords, zoom_start=10, tiles='openstreetmap')
small_df_sommuns = df_summons.iloc[::500]
for coords in list(zip(small_df_sommuns['Latitude'], small_df_sommuns['Longitude'])):
    folium.Circle(coords, radius=20, fill=True, color="blue").add_to(m_1)

small_df_collisions = df_collisions.iloc[::5000][['LATITUDE', 'LONGITUDE', 'NUMBER OF PERSONS INJURED']]
small_df_collisions = small_df_collisions.dropna().reset_index(drop=True)
for i, coords in enumerate(list(zip(small_df_collisions['LATITUDE'], small_df_collisions['LONGITUDE']))):
    folium.Circle(coords, radius=float(100 * (small_df_collisions['NUMBER OF PERSONS INJURED'][i]) + 1),
                  fill=True, color="red").add_to(m_1)
m_1
```
